### 이건 gradient가 아니라 선형회귀에서 도출해낸 a와 b를 활용해 구하는 방식

```python
import numpy as np
```

```python
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])
```

```python
mx = np.mean(x)
my = np.mean(y)
```

```python
mx, my
```

image.png

```python
# 평균의 제곱
# 이건 분모 구하는 것
divisor = sum([(i - mx)**2 for i in x])
divisor
```

```python
# 이건 분자 구하는 것
def top(x, mx, y, my):
  d = 0
  for i in range(len(x)):
    d += (x[i] - mx) * (y[i] - my)
  return d
```

```python
dividend = top(x, mx, y, my)
dividend
```

```python
a = dividend / divisor
a # 기울기
```

```python
b = my - (mx* a)
b # y절편 b
```

# 평균 제곱 오차

파이썬 코딩으로 확인하는 평균 제곱 오차

```python
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

fake_a = 3
fake_b = 76
```

```python
def predict(x):
  return fake_a * x + fake_b
```

```python
predict_result = []
```

```python
for i in range(len(x)):
  predict_result.append(predict(x[i]))
  print("공부시간 =%.f, 실제점수 = %.f, 예측점수 = %.f" %(x[i], y[i], predict(x[i])))
```

```python
predict_result
```

```python
n = len(x)
def mse(y, y_pred):
  return (1/n) * sum((y - y_pred)**2 )
```

```python
print("평균 제곱 오차: " + str(mse(y, predict_result)))
```
