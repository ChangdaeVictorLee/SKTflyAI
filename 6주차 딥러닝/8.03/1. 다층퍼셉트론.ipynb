```python
w11 = np.array([-2, -2])
w12 = np.array([2, 2])
w2 = np.array([1, 1])
b1 = 3
b2 = -1
b3 = -1
```

```python
# 퍼셉트론 알고리즘
# 0과 1중에서 값을 출력
def MLP(x, w, b):
  y = np.sum(w * x) + b # 하나의 노드의 값이 0초과면 1 아니면 0
  if y <= 0:
    return 0
  else:
    return 1
```

```python
# nand 게이트
# 모든 입력이 참일 때에만 거짓인 출력을 내보내는 논리 회로
def NAND(x1, x2):
  return MLP(np.array([x1, x2]), w11, b1)
```

```python
# OR 게이트
def OR(x1, x2):
  return MLP(np.array([x1, x2]), w12, b2)
```

```python
# AND 게이트
def AND(x1, x2):
  return MLP(np.array([x1, x2]), w2, b3)
```

```python
# XOR 게이트
def XOR(x1, x2):
  return AND(NAND(x1, x2), OR(x1, x2))
```

```python
for x in [(0,0), (1,0), (0,1), (1,1)]:
  y = XOR(x[0], x[1])
  print("입력 값: " + str(x) + "출력값: "+ str(y))
```
