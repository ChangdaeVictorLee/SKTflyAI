```python
import newspaper
import matplotlib.pyplot as plt
from wordcloud import WordCloud

font_path = './sample/NanumGothic.ttf'
wordcloud = WordCloud(font_path = font_path, background_color = 'white', width = 400, height =400)

link = 'https://www.edaily.co.kr/news/read?newsId=01676086632394520&mediaCodeNo=257&OutLnkChk=Y'

article = newspaper.Article(link, language = 'ko')
article.download()
article.parse()

#print(article.text)

wordcloud = wordcloud.generate(article.text)
fig = plt.figure(figsize = (6,6))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
fig.savefig("wordcloud.png")
```

# 한글 포인트

```python
import matplotlib.pyplot as plt
 
# 한글 폰트 사용을 위해서 세팅
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
fig = plt.figure(figsize = (10, 4))
ax1 = fig.add_subplot(111)
ax1.plot(df.index, df['Close'], color='b', label='주가')
# X축을 공유해서 Y축을 2개 사용하는 설정
ax2 = ax1.twinx()
ax2.bar(df.index, df['Volume'], color='g', label = '거래총액', width=2)

# 축과 축레이블 설정
ax1.set_xlabel('년월')
ax1.set_yticks([i * 2000 for i in range(5)])
ax1.set_ylabel('주가')
ax2.set_yticks([i * 5000 for i in range(5)])
ax2.set_ylabel('거래총액')
# 그래프 타이틀 설정
ax1.set_title('주가와 거래총액')
# 범래설정
# 두개가 깔리기에 위치를 loc1, 2 로 변경해줌
ax1.legend(loc=1)
ax2.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('ggplot')
# 입력값 생성
np.random.seed(2)
x = np.arange(1, 101)
y = 4 * x * np.random.rand(100)
# 산포도 그래프 그리기
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x, y)
plt.show()
```

```python
import os
import pandas as pd

df = pd.read_csv('./sample/anime_master.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(df['members'], df['rating'], alpha = 0.5)
plt.show()
```

```python
df.loc[df['members']>=800000, ['name','members']]
```

```python
df.loc[(df['members']>=600000) & (df['rating']>8.5), ['name','rating']]
```

```python
types = df['type'].unique()
types
```

```python
fig = plt.figure(figsize = (10,5))
ax = fig.add_subplot(111)

# 돌아가면서 타입이 맞을때만 뽑아내기
for t in types:
    x = df.loc[df['type'] == t, 'members']
    y = df.loc[df['type'] == t, 'rating']
    ax.scatter(x,y,alpha = 0.5, label =t)
    break
```
