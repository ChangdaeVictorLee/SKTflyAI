```python
import newspaper
import matplotlib.pyplot as plt
from wordcloud import WordCloud

font_path = './sample/NanumGothic.ttf'
wordcloud = WordCloud(font_path = font_path, background_color = 'white', width = 400, height =400)

link = 'https://www.edaily.co.kr/news/read?newsId=01676086632394520&mediaCodeNo=257&OutLnkChk=Y'

article = newspaper.Article(link, language = 'ko')
article.download()
article.parse()

#print(article.text)

wordcloud = wordcloud.generate(article.text)
fig = plt.figure(figsize = (6,6))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
fig.savefig("wordcloud.png")
```

```python
import os
import pandas as pd

df = pd.read_csv('./sample/anime_master.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(df['members'], df['rating'], alpha = 0.5)
plt.show()
```

```python
df.loc[df['members']>=800000, ['name','members']]
```

```python
df.loc[(df['members']>=600000) & (df['rating']>8.5), ['name','rating']]
```

```python
types = df['type'].unique()
types
```

```python
fig = plt.figure(figsize = (10,5))
ax = fig.add_subplot(111)

#    
for t in types:
    x = df.loc[df['type'] == t, 'members']
    y = df.loc[df['type'] == t, 'rating']
    ax.scatter(x,y,alpha = 0.5, label =t)
    break
```
