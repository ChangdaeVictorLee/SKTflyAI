## 원 코드

```python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://dhlottery.co.kr/common.do?method=main')
soup = BeautifulSoup(res.content, 'html.parser')
result = soup.select('.ball_645')
for num in result:
    print(num.text,end="번 ")
```
