```python
import os
import requests
from bs4 import BeautifulSoup as bs

url = 'https://comic.naver.com/webtoon/detail?titleId=710761&no=197&weekday=thu'
html = requests.get(url).text
soup = bs(html, 'html.parser')
result = soup.select('.wt_viewer img')


# 디렉토리를 만듦
# 폴더가 없으면 만들겠다
if not(os.path.isdir('./webtoon')):
    os.makedirs(os.path.join('./webtoon'))

    
i = 1 
for tag in result:
    print(tag['src'])
    img_url = tag['src']
    headers = {'Referer':img_url}
    img_data = requests.get(img_url, headers = headers)
    # zfill은 없는 공간을 0으로 채우는 것
    # 끝에서 -4는 자기의 확장자를 추출해옴.(jpg)
    save_img = "./webtoon/" + str(i).zfill(3) +img_url[-4:]
    print(save_img)
    with open(save_img, 'wb') as f:
        f.write(img_data.content)
    i += 1
```

# 요 아래부터 성공한 코드

```python
import os
import requests
from bs4 import BeautifulSoup as bs

url = 'https://www.google.com/search?q=%EA%B0%95%EC%95%84%EC%A7%80&source=lnms&tbm=isch&sa=X&ved=2ahUKEwj3tez_7vf4AhU8mlYBHbv1BVYQ_AUoAXoECAIQAw&biw=1536&bih=714&dpr=1.25'
html = requests.get(url).text
soup = bs(html, 'html.parser')

result = soup.select('.yWs4tf')
```

```python
result[0]['src']
```

```python
pic_list = []

for i in range(len(result)):
    pic_list.append(result[i]['src'])
```

```python
pic_list
```

```python
if not(os.path.isdir('./puppy')):
    os.makedirs(os.path.join('./puppy'))

i = 1 
for tag in pic_list:
    headers = {'Referer':tag}  
    img_data = requests.get(tag, headers =headers)
    save_img = "./puppy/" + str(i).zfill(3)+ ".jpg"
    print(save_img)
    with open(save_img, 'wb') as f:
        f.write(img_data.content)
    i += 1
```
