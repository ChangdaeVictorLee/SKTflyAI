```python
# 정렬을 episodes 내림차순으로 바꾸고 싶음
from pandasql import sqldf

dfsql = lambda q:sqldf(q,globals())
result = dfsql("select name,episodes from df where type='Movie' order by episodes desc")
result
# json으로 저장할 때 속성은 table
result.to_json('sample.json', orient='table')
```
