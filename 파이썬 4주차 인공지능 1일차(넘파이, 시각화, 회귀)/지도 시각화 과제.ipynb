```python
import pandas as pd
import matplotlib.pyplot as plt
import folium
```

```python
df = pd.read_csv("서울시 응급실 위치 정보.csv", encoding = 'cp949')
```

```python
df.head()
```

```python
df[['병원경도', '병원위도']]
```

```python
seoul_map = folium.Map(location=[37.55,126.98], tiles='cartodbpositron', zoom_start=11)
seoul_map
```

```python
seoul_map = folium.Map(location=[37.55,126.98], tiles='Stamen Terrain', zoom_start=12)
# 대학교 위치정보를 Marker로 표시
for name, lat, lng, phone in zip(df.기관명, df.병원위도, df.병원경도, df.응급실전화):
    folium.Marker([lat, lng], popup= (name, phone)).add_to(seoul_map)
seoul_map
```

```python
seoul_map.save('./seoul.html')
```
