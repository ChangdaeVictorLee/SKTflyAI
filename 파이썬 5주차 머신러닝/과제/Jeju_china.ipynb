```python
china.head(10)
```

```python
china.방문인구 = china.방문인구.apply(np.round)
```

```python
china.방문인구 = china.방문인구.apply(int)
```

```python
china.head()
```

```python
pd.get_dummies(china['읍면동명'])
```

```python
pd.get_dummies(china['업종'])
```

```python
china_onehot = pd.concat([pd.get_dummies(china['업종']), pd.get_dummies(china['읍면동명']), china[['방문인구', '이용자수', '이용금액']]], axis=1)
china_onehot
```

```python
china_onehot.iloc[:,-1]
```

```python
import matplotlib.pylab as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(china_onehot.iloc[:,:-1], china_onehot.iloc[:,-1], test_size=0.2, random_state=0)
# 선형 회귀 모델로 학습을 수행한다.
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
pred = model.predict(X_test)

r2_score(y_test, pred)
```

## OLS 진행

```python
import statsmodels.api as sm
from scipy import stats

X = china_onehot.iloc[:,:-1]
y = china_onehot.iloc[:,-1]

X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```
